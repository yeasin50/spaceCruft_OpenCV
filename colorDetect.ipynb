```python
import numpy as np
import cv2
```

```python
image =  cv2.imread(r'testImages/tb1.jpg')
```

```python
# define the list of boundaries
boundaries = [
    ([122, 822, 102], [245,237, 211]),# light On  
]
```

([17, 15, 100], [50, 56, 200])  
Here, we are saying that all pixels in our image that have a R >= 100, B >= 15, and G >= 17 along with R <= 200, B <= 56, and G <= 50 will be considered red.

```python
# loop over the boundaries
for (lower, upper) in boundaries:
	# create NumPy arrays from the boundaries
	lower = np.array([0,0,200], dtype = "uint8")
	upper = np.array(upper, dtype = "uint8")
	# find the colors within the specified boundaries and apply
	# the mask
	mask = cv2.inRange(image, lower, upper)
	output = cv2.bitwise_and(image, image, mask = mask)
	# show the images
	cv2.imshow("images", np.hstack([image,output]))
	cv2.waitKey(0)
    
    
cv2.destroyAllWindows()
```
